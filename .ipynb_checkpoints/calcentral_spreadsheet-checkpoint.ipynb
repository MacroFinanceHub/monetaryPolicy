```python
import os
import pandas as pd
import numpy as np
```

```python
path = '../calcentral/app/models'

folders = dict()

# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for folder in d:
        folders.append(os.path.join(r, folder))

for f in folders:
    print(f)
```

```python
rootDir = '../calcentral/app/models'

modelDict = dict()
for dirName, subdirList, fileList in os.walk(rootDir):
    strippedDirName = dirName.replace("../calcentral/app/models/", "")
    modelDict[strippedDirName] = []
    for fname in fileList:
        modelDict[strippedDirName].append(fname)
```

```python
modelDict
```

```python
import xlwt 
  
workbook = xlwt.Workbook()  
  
sheet = workbook.add_sheet("assignments") 
  
# Specifying style 
style = xlwt.easyxf('font: bold 1') 

index = 0
totalNum = 0

# Specifying column 
for key in modelDict:
    sheet.write(index, 0, key, style)
    length_of_directory = len(modelDict[key])
    sheet.write(index, 1, length_of_directory)
    index = index + 1
    totalNum = totalNum + length_of_directory
    for i in range(length_of_directory):
        item = modelDict[key][i]
        sheet.write(index, 0, item)
        index = index + 1

workbook.save("calcentral.xls") 
```
